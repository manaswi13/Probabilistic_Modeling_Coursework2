```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
```

```python
# Load and select data
data = pd.read_excel('Fungus.xlsx', sheet_name='Sheet2', header=None)
selected_data = data.iloc[76:276]
X = selected_data.to_numpy()
```

```python
# Fit GMM
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=42)
gmm.fit(X)
labels = gmm.predict(X)
```

```python
# Plot data with cluster assignments
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='.', s=10)
plt.xlabel('x (relative to farm center)')
plt.ylabel('y (relative to farm center)')
plt.title('GMM Clustering of Fungus Locations (Entries 76275)')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('gmm_plot.png')
plt.show()

# Print parameters
print('Mixing Coefficients:', gmm.weights_)
print('Means:', gmm.means_)
print('Covariances:', gmm.covariances_)
```
