```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from Excel
data = pd.read_excel('Fungus.xlsx', sheet_name='Sheet2', header=None)

# Select 200 entries starting from index 76 (0-based)
start_idx = 76
selected_data = data.iloc[start_idx:start_idx+200]

# Extract x and y coordinates
x = selected_data[0]
y = selected_data[1]

# Create scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(x, y, marker='.', color='blue', s=10)
plt.xlabel('x (relative to farm center)')
plt.ylabel('y (relative to farm center)')
plt.title('Scatter Plot of Fungus Locations (Entries 76â€“275)')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('fungus_plot.png')
plt.show()
```
